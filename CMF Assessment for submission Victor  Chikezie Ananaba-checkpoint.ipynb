```python
#Black Scholes Model
```

```python
def XRP_euro_option_bsm(S, K, T, r, vol, XRP_payoff):
    
    #S: spot price
    #K: strike price
    #T: time to maturity
    #r: risk free rate
    #vol: volatility of underlying asset
    #payoff: call or put
    d1 = (np.log(S / K) + (r + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    d2 = (np.log(S / K) + (r - 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    if XRP_payoff == "call":
        XRP_ov = S * si.norm.cdf(d1, 0.0, 1.0) - K * np.exp(-r * T) * si.norm.cdf(d2, 0.0, 1.0)
    elif XRP_payoff == "put":
        XRP_ov = - S * si.norm.cdf(-d1, 0.0, 1.0) + K * np.exp(-r * T) * si.norm.cdf(-d2, 0.0, 1.0)
    
    return XRP_ov
```

```python
#payoff: call
S = XRP['Adj Close'][-1]
K = 0.80
T = 1
r = 0.05
XRP_vol
XRP_bsm_call = XRP_euro_option_bsm(S, K, T, r, XRP_vol, 'call')
print("The fair value of Black Scholes Model European call option is: ${:.8}".format(XRP_bsm_call))
```

```python
def delta(S, K, T, r, q, vol, payoff):
    
    d1 = (np.log(S / K) + (r - q + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    if payoff == "call":
        delta = np.exp(-q * T) * si.norm.cdf(d1, 0.0, 1.0)
    elif payoff == "put":
        delta =  - np.exp(-q * T) * si.norm.cdf(-d1, 0.0, 1.0)
    
    return delta
```

```python
S = XRP['Adj Close'][-1]
K = 0.80
T = 1
r = 0.05
q = 0.0
vol = 0.80
d= delta(S, K, T, r, q, vol, 'call')
print('The value of Delta is', d.round(4),'.','If the token price increase 1 dollar, then the value of the option will increase $',d.round(4),'.')
```

```python
S = np.linspace(50,150,11)
Delta_Call = np.zeros((len(S),1))
for i in range(len(S)):
    Delta_Call [i] = delta(S[i], K, T, r, q, XRP_vol, 'call')
```

```python
#2D figure
fig = plt.figure()
plt.plot(S, Delta_Call, '-')
plt.grid()
plt.xlabel('Token Price')
plt.ylabel('Delta')
plt.title('Delta')
plt.legend(['Delta for Call'])
```

```python
def gamma(S, K, T, r, q, vol, payoff):
    
    d1 = (np.log(S / K) + (r - q + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    gamma = np.exp(- q * T) * si.norm.pdf(d1, 0.0, 1.0) / (vol * S * np.sqrt(T))
    
    return gamma
```

```python
def speed(S, K, T, r, q, vol, payoff):
    
    d1 = (np.log(S / K) + (r - q + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    speed = - np.exp(-q * T) * si.norm.pdf(d1, 0.0, 1.0) / ((vol **2) * (S**2) * np.sqrt(T)) * (d1 + vol * np.sqrt(T))
    
    return speed
```

```python
S = 0.86
K = 0.9
T = 1
r = 0.05
q = 0.0
vol = 0.80
speed(S, K, T, r, q, vol, 'call')
```

```python
def theta(S, K, T, r, q, vol, payoff):
    
    d1 = (np.log(S / K) + (r - q + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    d2 = (np.log(S / K) + (r - q - 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    if payoff == "call":
        theta = vol * S * np.exp(-q * T) * si.norm.pdf(d1, 0.0, 1.0) / (2 * np.sqrt(T)) - q * S * np.exp(-q * T) * si.norm.cdf(d1, 0.0, 1.0) + r * K * np.exp(-r * T) * si.norm.cdf(d2, 0.0, 1.0)
    elif payoff == "put":
        theta = vol * S * np.exp(-q * T) * si.norm.pdf(-d1, 0.0, 1.0) / (2 * np.sqrt(T)) - q * S * np.exp(-q * T) * si.norm.cdf(-d1, 0.0, 1.0) + r * K * np.exp(-r * T) * si.norm.cdf(-d2, 0.0, 1.0)
    
    return theta
```

```python
S = 0.86
K = 0.9
T = 1
r = 0.05
q = 0.0
vol = 0.80
theta(S, K, T, r, q, vol, 'call')
```

```python
T = np.linspace(0.25,3,12)
Theta_Call = np.zeros((len(T),1))
Theta_Put = np.zeros((len(T),1))
for i in range(len(T)):
    Theta_Put [i] = theta(S, K, T[i], r, q, vol, 'call')
```

```python
fig = plt.figure()
plt.plot(T, Theta_Put, '-')
plt.grid()
plt.xlabel('Time to Expiry')
plt.ylabel('Theta')
plt.title('Theta')
plt.legend(['Theta for Call'])
```

```python
#Vega
```

```python
def vega(S, K, T, r, q, vol, payoff):
    
    d1 = (np.log(S / K) + (r - q + 0.5 * vol ** 2) * T) / (vol * np.sqrt(T))
    vega = S * np.sqrt(T) * np.exp(-q * T) * si.norm.pdf(d1, 0.0, 1.0)
    
    return vega
```

```python
S = 0.86
K = 0.9
T = 1
r = 0.05
q = 0.0
vol = 0.80
vega(S, K, T, r, q, vol, 'call')
```

```python
vol = np.linspace(0.1,0.3,25)
Vega = np.zeros((len(vol),1))
for i in range(len(vol)):
    Vega [i] = vega(S, K, T, r, q, vol[i],'call')
```

```python
fig = plt.figure()
plt.plot(vol, Vega, '-')
plt.grid()
plt.xlabel('Volatility')
plt.ylabel('Vega')
plt.title('Vega')
plt.legend(['Vega for Call'])
```
